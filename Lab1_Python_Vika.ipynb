```python
from random import randint

# Приклад з інструкції
A = randint(0, 1)
print(f"Значить А={A}" if A else "Але може бути що А={}".format(A))

# --- Власний приклад (Перевірка ігрового статусу) ---
health = randint(0, 100)
print(f"\nТвоє здоров'я: {health}%")

if health > 70:
    print("Статус: Ти в повній бойовій готовності!")
elif health > 20:
    print("Статус: Потрібно бути обережною, здоров'я менше норми.")
else:
    print("Статус: Терміново шукай аптечку!")
```
